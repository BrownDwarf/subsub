```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import seaborn as sns
import pandas as pd
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

The "good" orders right now are 106, 109, 116, and 119. We're re-running some orders to hopefully get rid of a bimodal distribution.  

Let's also try labeling some lines from the Arcturus line list

```python
arct = pd.read_csv('/Users/ngosnell/Rayner2009_Arcturuslines.csv')
```

```python
arct.wl = arct.wl * 10000
arct.head()
```

```python
df = pd.read_csv('../sf/m106/output/mix_emcee/run01/models_ff-05_50_95.csv')
```

```python
matchids = arct.loc[(arct['wl'] >= df.wl.min()) & (arct['wl'] <= df.wl.max())]
ids = list(matchids.id)
idwl = np.array(matchids.wl)
```

```python
plt.figure(figsize=(10,6))
plt.step(df.wl, df.data, label='data')
plt.plot(df.wl, df.model_comp50, label='Composite')
plt.plot(df.wl, df.model_cool50, label='Cool')
plt.plot(df.wl, df.model_hot50, label='Hot')
for i in range(len(ids)):
    plt.annotate(ids[i], (idwl[i], 0.4), xycoords='data')
#plt.annotate(arct.id[444], (arct.wl[444], 0.4), xycoords = 'data')
plt.legend(loc='best');
```
